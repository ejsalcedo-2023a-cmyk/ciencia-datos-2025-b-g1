1. Definición de Ciencia de Datos: es un campo multidisciplinario que combina matemáticas, estadística, programación e inteligencia artificial para extraer información valiosa y conocimientos de grandes volúmenes de datos. Su objetivo es usar estos insights para la toma de decisiones informadas, hacer predicciones y resolver problemas complejos, cubriendo todo el ciclo de vida de los datos, desde su obtención hasta la comunicación de resultados. 

2. Roles en la Ciencia de Datos

-Ingeniero de Datos, enfocado en la infraestructura y el flujo de datos.
-Analista de Datos, que interpreta datos históricos para el presente.
-Científico de Datos, que combina análisis y predicción para el futuro.
-Ingeniero de Machine Learning, especializado en la puesta en producción de modelos.
-Arquitecto de Datos, que diseña la estructura general.
-Analista de Negocio, que aporta la perspectiva empresarial.

3. ¿Qué son los dataset?

Son un conjunto de datos estructurados, como una tabla, donde cada fila es una entrada individual y cada columna es una variable o característica, permitiendo su análisis, procesamiento y manipulación para encontrar patrones, hacer cálculos o generar visualizaciones. Su objetivo principal es facilitar el trabajo con la información, siendo la "materia prima" para el análisis de datos y el aprendizaje automático.

Un DataFrame es una estructura de datos bidimensional, con filas y columnas, que organiza la información de forma similar a una hoja de cálculo o tabla de base de datos. Esta estructura es fundamental en lenguajes como Python (donde es un objeto de la librería Pandas) y R, ya que permite almacenar y manipular datos de diversos tipos de manera intuitiva para análisis, filtrado y visualización. 

5. Generar una gráfica con el anterior `df`

```python
# Importar la librería 
import matplotlib.pyplot as plt 

# Datos de ejemplo 
x = [1, 2, 3, 4, 5] 
y = [2, 4, 7, 11, 16]

# Crear la gráfica 
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Crecimiento') 

# Personalizar la gráfica 
plt.title("Ejemplo de gráfica en Notebook") 
plt.xlabel("Eje X") 
plt.ylabel("Eje Y") 
plt.legend() 
plt.grid(True) 

# Mostrar la gráfica 
plt.show()
```
